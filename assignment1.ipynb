```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = 'img/lena.png'
img = cv2.cvtColor(cv2.imread(img_path),cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
height,width= img.shape[:2] 
print(height,width)
```

### 线性点运算

```python
out1=np.zeros((height,width),np.uint8)
out2=np.zeros((height,width),np.uint8)
out3=np.zeros((height,width),np.uint8)

for i in range(height):
    for j in range(width):
        out1[i,j]=img[i,j]*2+2
        out2[i,j]=img[i,j]*0.5+2
        out3[i,j]=img[i,j]*(-2)+2

plt.figure(figsize=(12, 10))
plt.subplot(1,4,1)
plt.imshow(img, cmap="gray")
plt.title("origin image")
plt.subplot(1,4,2)
plt.imshow(out1, cmap="gray")
plt.title("s=2r+2")
plt.subplot(1,4,3)
plt.imshow(out2, cmap="gray")
plt.title("s=0.5r+2")
plt.subplot(1,4,4)
plt.imshow(out3, cmap="gray")
plt.title("s=-2r+2")
plt.show()
```

### 分段线性点运算

定义分段函数

```python
def transfunc(x):
    if x<=40:
        y=0.1*x
    elif x>100:
        y=0.1*x
    else:
        y=10*x+2
    return y
```

运算

```python
out4=np.zeros((height,width),np.uint8)

for i in range(height):
    for j in range(width):
        out4[i,j]=transfunc(img[i,j])

plt.figure(figsize=(10, 8))
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("origin image")
plt.subplot(1,2,2)
plt.imshow(img, cmap="gray")
plt.title("transfered image")
plt.show()
```

### 非线性点运算

```python
out5=np.zeros((height,width),np.uint8)
out6=np.zeros((height,width),np.uint8)
out7=np.zeros((height,width),np.uint8)

for i in range(height):
    for j in range(width):
        out5[i,j]=10*np.log2(img[i,j]+10) #对数变换
        out6[i,j]=img[i,j]^5 #幂次变换

plt.figure(figsize=(12, 10))
plt.subplot(1,3,1)
plt.imshow(img, cmap="gray")
plt.title("origin image")
plt.subplot(1,3,2)
plt.imshow(out5, cmap="gray")
plt.title("s=10log2(r+10)")
plt.subplot(1,3,3)
plt.imshow(out6, cmap="gray")
plt.title("s=r^5")
plt.show()
```
