```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = 'img/test.jpg'
img = cv2.cvtColor(cv2.imread(img_path),cv2.COLOR_BGR2RGB)
plt.imshow(img)
height,width= img.shape[:2] 
print(height,width)
```

### 

```python
dx=100
dy=100
M1=np.array([[1,0,dx],
            [0,1,dy]],
            dtype=float)
dst1=cv2.warpAffine(img,M1,dsize=(width,height))
plt.imshow(dst1)
plt.show()
```

### 

#### 

```python
M2=np.array([[-1,0,width],
            [0,1,0]],
            dtype=float)
dst2=cv2.warpAffine(img,M2,dsize=(width,height))
plt.imshow(dst2)
plt.show()
```

#### 

```python
M3=np.array([[1,0,0],
            [0,-1,height]],
            dtype=float)
dst3=cv2.warpAffine(img,M3,dsize=(width,height))
plt.imshow(dst3)
plt.show()
```

### 

```python
beta=np.pi/6
M4=np.array([[np.cos(beta),np.sin(beta),0],
             [-np.sin(beta),np.cos(beta),0]],
             dtype=float)
dst4=cv2.warpAffine(img,M4,dsize=(width,height))
plt.imshow(dst4)
plt.show()
```

### 

```python
temp1=cv2.warpAffine(dst1,M2,dsize=(width,height))#
out_img=cv2.warpAffine(temp1,M4,(width,height))#
plt.imshow(out_img)
plt.show()
```
